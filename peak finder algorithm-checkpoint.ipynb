```python
def find_nearest_points(dataset1, point, k=6):
    """
    Finds the k nearest points to a given point in a dataset of 2D points.
    
    Parameters:
        dataset (numpy.ndarray): Array of shape (n, 2) representing the dataset of 2D points.
        point (tuple or list): The given point (x, y).
        k (int): Number of nearest neighbors to find. Default is 5.
        
    Returns:
        numpy.ndarray: Array of the k nearest points, sorted by distance.
    """
    
    dataset=np.array(dataset1)
    # Compute Euclidean distances
    distances = np.linalg.norm(dataset - point, axis=1)
    
    # Get indices of the k smallest distances
    nearest_indices = np.argsort(distances)[:k]
    
    # Retrieve the k nearest points
    nearest_points = dataset[nearest_indices]
    
    return nearest_points
```

```python
import numpy as np
import matplotlib.pyplot as plt
from math import acos, pi

def generate_points():
    """Generate a dataset of 2D points, including a grid and circular distribution."""
    points = []
    
    # Add points in a grid
    x = 0
    while x <= 1:
        y = 0
        while y <= 1:
            points.append([x, y])
            y += 0.1
        x += 0.1
        
    return np.array(points)

def find_nearest_points(dataset, point, k=6):
    """Find the k nearest points to a given point in a dataset of 2D points."""
    distances = np.linalg.norm(dataset - point, axis=1)
    nearest_indices = np.argsort(distances)[:k]
    return dataset[nearest_indices]

def assign_random_values_with_peaks(points, peaks_indices, random_seed=None):
    """Assign random values to points, boosting values at peak indices."""
    if random_seed is not None:
        np.random.seed(random_seed)
    
    random_values = np.random.random(len(points))
    random_values[peaks_indices] += 500
    return random_values

def inside_poly(point, polygon):
    """
    Check if a point is inside a polygon using the ray-casting algorithm.

    Parameters:
        point (tuple): The (x, y) coordinates of the point.
        polygon (list of tuples): A list of (x, y) coordinates representing the vertices of the polygon.
                                  The vertices should be ordered either clockwise or counterclockwise.
    
    Returns:
        bool: True if the point is inside the polygon, False otherwise.
    """
    x, y = point
    n = len(polygon)
    inside = False

    for i in range(n):
        x1, y1 = polygon[i]
        x2, y2 = polygon[(i + 1) % n]  # Wrap around to the first vertex

        # Check if the point is between the y-coordinates of the edge
        if min(y1, y2) < y <= max(y1, y2):
            # Compute the x-coordinate of the intersection of the ray with the edge
            x_intersect = x1 + (y - y1) * (x2 - x1) / (y2 - y1)
            # Flip the inside status if the ray crosses the edge
            if x < x_intersect:
                inside = not inside

    return inside

def find_peaks(points, values, k=6):
    """Find maxima using 2D quadratic interpolation."""
    peaks = []
    
    for i in range(len(points)):
        try:
            p0 = points[i]
            nearest = find_nearest_points(points, p0, k=k)
        
            M = np.zeros([k, k])
            B = np.zeros(k)
        
            for j in range(k):
                M[j][0] = 1
                M[j][1] = nearest[j][0]
                M[j][2] = nearest[j][1]
                M[j][3] = nearest[j][0] * nearest[j][1]
                M[j][4] = nearest[j][0] ** 2
                M[j][5] = nearest[j][1] ** 2
                ind = np.where((points == nearest[j]).all(axis=1))[0][0]
                B[j] = values[ind]
        
            y1 = np.linalg.solve(M, B)

            v1 = [-y1[1], -y1[2]]
            M1 = [[y1[4], y1[3]], [y1[3], y1[5]]]
            max_point = np.linalg.solve(M1, v1)

            second_der_x = 2 * y1[4]
            second_der_y = 2 * y1[5]
        
            if inside_poly(max_point, nearest):
                peaks.append(max_point)
                
        except:
            pass

    return np.array(peaks)

# Main Program
points = generate_points()
peak_indices = np.array([1, 3, 45, 66, 100])  # Example peak indices
values = assign_random_values_with_peaks(points, peak_indices)

# Find peaks
peaks = find_peaks(points, values)

# Plot points, values, and peaks
plt.scatter(points[:, 0], points[:, 1], c=values, cmap="viridis", label="Points")
plt.colorbar(label="Assigned Values")
plt.scatter(peaks[:, 0], peaks[:, 1], color="red", label="Peaks", zorder=5)
plt.legend()
plt.title("2D Points with Peaks")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()

```

```python
from scipy.spatial import ConvexHull

def find_peaks(points, values, k=6):
    """Find maxima using 2D quadratic interpolation."""
    peaks = []
    
    for i in range(len(points)):
        try:
            p0 = points[i]
            nearest = find_nearest_points(points, p0, k=k)
        
            M = np.zeros([k, k])
            B = np.zeros(k)
        
            for j in range(k):
                M[j] = [1, nearest[j][0], nearest[j][1], 
                        nearest[j][0] * nearest[j][1], 
                        nearest[j][0] ** 2, nearest[j][1] ** 2]
                ind = np.argmin(np.linalg.norm(points - nearest[j], axis=1))
                B[j] = values[ind]
        
            y1 = np.linalg.solve(M, B)

            v1 = [-y1[1], -y1[2]]
            H = np.array([[2 * y1[4], y1[3]], [y1[3], 2 * y1[5]]])
            max_point = np.linalg.solve(H, v1)

            if np.linalg.det(H) > 0 and H[0, 0] < 0:  # Local max condition
                hull = ConvexHull(nearest)
                inside = all(np.dot(eq[:-1], max_point) + eq[-1] <= 0 for eq in hull.equations)
                if inside:
                    peaks.append(max_point)
                
        except np.linalg.LinAlgError:
            continue

    return np.array(peaks)

points = generate_points()
peak_indices = np.array([1, 3, 45, 66, 100])  # Example peak indices
values = assign_random_values_with_peaks(points, peak_indices)

# Find peaks
peaks = find_peaks(points, values)

# Plot points, values, and peaks
plt.scatter(points[:, 0], points[:, 1], c=values, cmap="viridis", label="Points")
plt.colorbar(label="Assigned Values")
plt.scatter(peaks[:, 0], peaks[:, 1], color="red", label="Peaks", zorder=5)
plt.legend()
plt.title("2D Points with Peaks")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import ConvexHull

def generate_2d_points():
    """Generate a 2D grid of points with some added circular distribution."""
    points = []
    
    # Generate a grid of points
    x = np.linspace(0, 1, 11)
    y = np.linspace(0, 1, 11)
    for i in x:
        for j in y:
            points.append([i, j])
    
    # Add points forming a circular distribution
    theta = np.linspace(0, 2 * np.pi, 30)
    r = 0.25
    circle_points = [[0.5 + r * np.cos(t), 0.7 + r * np.sin(t)] for t in theta]
    points.extend(circle_points)
    
    return np.array(points)

def assign_random_values(points, peak_indices=None, seed=None):
    """Assign random values to points, boosting peak indices."""
    np.random.seed(seed)
    values = np.random.random(len(points))
    
    if peak_indices is not None:
        for idx in peak_indices:
            values[idx] += 5  # Boost values at peaks
    return values

def find_nearest_neighbors(points, target, k=6):
    """Find the k nearest neighbors of a target point."""
    distances = np.linalg.norm(points - target, axis=1)
    nearest_indices = np.argsort(distances)[:k]
    return points[nearest_indices]

def check_convex_hull(max_point, neighbors):
    """Check if a point lies inside the convex hull of a set of points."""
    hull = ConvexHull(neighbors)
    equations = hull.equations
    return all(np.dot(eq[:-1], max_point) + eq[-1] <= 0 for eq in equations)

def find_peaks(points, values, k=6):
    """Find peaks using 2D quadratic interpolation."""
    peaks = []
    
    for i, point in enumerate(points):
        try:
            neighbors = find_nearest_neighbors(points, point, k)
            M = np.zeros((k, k))
            B = np.zeros(k)
            
            for j, neighbor in enumerate(neighbors):
                M[j] = [1, neighbor[0], neighbor[1], neighbor[0] * neighbor[1], 
                        neighbor[0] ** 2, neighbor[1] ** 2]
                B[j] = values[np.where((points == neighbor).all(axis=1))[0][0]]
            
            coeffs = np.linalg.solve(M, B)
            Hessian = np.array([[2 * coeffs[4], coeffs[3]], [coeffs[3], 2 * coeffs[5]]])
            gradient = np.array([-coeffs[1], -coeffs[2]])
            
            # Solve for the maximum point
            max_point = np.linalg.solve(Hessian, gradient)
            
            if np.linalg.det(Hessian) > 0 and Hessian[0, 0] < 0:  # Local max check
                if check_convex_hull(max_point, neighbors):
                    peaks.append(max_point)
        except np.linalg.LinAlgError:
            continue
    
    return np.array(peaks)

# Main program
points = generate_2d_points()
seed = 42
peak_indices = [50, 80, 120]  # Example indices for peaks
values = assign_random_values(points, peak_indices, seed=seed)

# Find peaks
peaks = find_peaks(points, values)

# Plotting
plt.figure(figsize=(10, 8))
plt.scatter(points[:, 0], points[:, 1], c=values, cmap="viridis", label="Data Points")
plt.colorbar(label="Values")
if len(peaks) > 0:
    plt.scatter(peaks[:, 0], peaks[:, 1], color="red", label="Detected Peaks", zorder=5)
plt.title("2D Peak Detection")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```
